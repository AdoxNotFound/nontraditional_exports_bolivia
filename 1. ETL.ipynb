A continuacion se define una funacion que permitir√° obtener los diferentes archivos de excel que estan en la pagina del ine y posteriormente unirlos dentro de un solo dataframe

```python
def read_web_xls(url):
  r = requests.get(url)
  open('temp.xls', 'wb').write(r.content)
  df = pd.read_excel('temp.xls')
  return df
```

### 2.Exploratory data analysis (EDA)

```python
df_ex.info()
```

al ver que no tenemos como tal datos nulos, se utiliza la siguiente funcion para contabilizar los datos diferentes para cada columna y poder identificar si el numero de id de las diferentes columans coinciden con el numero de descripciones.

```python
#we can check the diferent values for each column
list_col = df_ex.columns
new_table = dict()
for col in list_col:
  list_unique = df_ex[col].unique()
  length = len(list_unique)
  new_table[col] = length

reporte = pd.DataFrame.from_dict(new_table, orient='index')
reporte.rename(columns={0: 'count'}, inplace=True)
reporte
```
